```python
%matplotlib inline
import random
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
data   = []  
ch = []
with open("1.csv",'rb') as ifile:
        for line in ifile:
            line=line.decode('utf-8')
            data.append(line)
for i in data:
    t=[int(a) for a in i.split(",")]
    temp=[]
    temp.append(t[1])
    if t[0]==1:
        x=3+random.uniform(-2, 2)
        y=3+random.uniform(-2, 2)
    elif t[0]==2:
        x=-3+random.uniform(-2, 2)
        y=3+random.uniform(-2, 2)
    elif t[0]==3:
        x=-3+random.uniform(-2, 2)
        y=-3+random.uniform(-2, 2)
    else :
        x=3+random.uniform(-2, 2)
        y=-3+random.uniform(-2, 2)
    temp.append(x)
    temp.append(y)
    ch.append(temp)
plt.scatter([a[1] for a in ch],[a[2] for a in ch],c=[a[0] for a in ch],s=50)
```
